```python
#Creating the purpose column as categorical
categorical= list(Loanstudy['purpose'])
```

```python
#printing the columns which are created as purpose
categorical
```

```python
#Creating dummies for all the categorical columns
Dataframe = pd.get_dummies(data = Loanstudy, columns = ['purpose'], drop_first = True)
```

```python
#Printing the head of the final data
Dataframe.head()
```

```python
#length of the dataframe after EDA
Dataframe.shape
```

```python
#importing the train_test_split
from sklearn.model_selection import train_test_split
```

```python
#Setting the x and y variable
X = Dataframe.drop('not.fully.paid', axis = 1)
y = Dataframe['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2 , random_state= 101)
```

```python
#
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
print('Accuracy of LDA classifier on training set: {:.2f}'
     .format(lda.score(X_train, y_train)))
print('Accuracy of LDA classifier on test set: {:.2f}'
     .format(lda.score(X_test, y_test)))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(X_train, y_train)
LR.predict(X_test)
round(LR.score(X_test,y_test), 4)
```

```python
from sklearn import svm
```

```python
SVM = svm.SVC(decision_function_shape="ovo").fit(X_train, y_train)
SVM.predict(X_test)
round(SVM.score(X_test, y_test), 4)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(logreg.score(X_train, y_train)))
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(logreg.score(X_test, y_test)))
```

```python
#importing the decisiontreeclassifier 
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
#Fitting the decision tree 
dtree.fit(X_train, y_train)
```

```python
#Predicting the decision tree variable 
y_predict = dtree.predict(X_test)
```

```python
#Importing the confusion matrix library and classification report 
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#Printing the classification report
print(classification_report(y_test, y_predict))
```

```python
print(confusion_matrix(y_test, y_predict))
```


We can look at the classification report to get an understanding on the predictions. Looking at the precision and recall we can come to the conclusion that the model was able to predict accurately for 86 percent of the 0's and 21% od the 1's.However, the loan company is interested in the people who do not pay the loan or the defaulters.From the recall we can identify that 23 percent of the people did are defaulters from the test data.

```python
#Importing the random forest regressor classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
rforest = RandomForestClassifier(n_estimators = 300)
```

```python
#Fitting the random forest model to the data
rforest.fit(X_train, y_train)
```

```python
y_predict = rforest.predict(X_test)
```

```python
print(classification_report(y_test,y_predict))
```

```python
print(confusion_matrix(y_test,y_predict))
```

Looking at the two models we can come to a conclusion that random forest has a higher accuracy, However, accuracy alone cannot be used to correctly predict. Random forest has a poor recall which is about 2 percent, Only 2 percent of people failed to pay back their loans which we understood from the test data.

Therefore to choose the model we have to check which parameter we want to optimize and based on that choose the model.
